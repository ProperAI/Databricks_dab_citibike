```python
#reference fields
select * from system.billing.usage
order by usage_start_time desc
limit 3;
```

```python

select workspace_id, sku_name, usage_date,
round(usage_quantity,3) usage_qty, billing_origin_product as product, usage_type,
usage_metadata
from system.billing.usage
where usage_date > '2016-04-16' 
order by usage_start_time desc
limit 20
```

```python
# total costs by product and usage
select billing_origin_product as product,
count(*) as instances,
round(sum(usage_quantity),2) usage_qty, 
billing_origin_product as product, 
usage_type
from system.billing.usage
where usage_date > '2016-04-16' 
group by 1, 4, 5
order by usage_qty desc
limit 20
```
