```python
import dlt
from pyspark.sql.functions import col, unix_timestamp, to_date
```

```python
@dlt.table(
    coment="Silver layer: clean and enrich Citi data."
)

def silver_jc_citibike():
    # Read bronze dlt table
    df_bronze = dlt.read("bronze_jc_citibike")

    df_silver = (
        df_bronze
            .withCoumn("trip_duration_mins", 
                    #    (unix_timestamp(col("ended_at"))- unix_timestamp(col("started_at")))/60
                       (col("ended_at").cast("long")-col("started_at").cast("long"))/60
                       )
            .withColumn("trip_start_date" , todate(col("started_at"))
                        )
            .select("ride_id",
            "trip_start_date",
            "started_at",
            "ended_at",
            "start_station_name",
            "end_station_name",
            "trip_duration_mins")
    )

    return df_silver
```
